**Count the number of unique tokens in the text.**

```python
text = "Soheli, a vibrant and determined individual, embraces life with a zest for adventure! üåü Born on 25/07/2001, she defies expectations and embodies the spirit of resilience. üéâ A self-proclaimed explorer, she doesn't let obstacles hinder her path, turning challenges into opportunities. üí™ Don't underestimate her, for she is a force to be reckoned with! üöÄ Oh, did I mention her email is sohelipaul25@gmail.com? Feel free to drop her a message for a chat about life's wonders. If you want to catch up, how about meeting on the 10/02/2022 or the 05/09/2023, or 15/10/2022 ? üóìÔ∏è And if you're feeling really adventurous, give her a call at 7001539043 or try reaching her on 9876543210 or maybe 91234567822, an invalid number she mistakenly shared once! üìû #SoheliAdventures #BornToExplore #GirlPower #DontHoldBack"
# Tokenization
tokens = text.split()
unique_tokens = set(tokens)

# Counting unique tokens
num_unique_tokens = len(unique_tokens)
print(f"Number of unique tokens: {num_unique_tokens}")
```

**Count the number of punctuation characters and remove punctuations.**

```python
import string

# Counting punctuation characters
num_punctuations = sum(1 for char in text if char in string.punctuation)
print(f"Number of punctuation characters: {num_punctuations}")

# Removing punctuations
text_without_punctuations = text.translate(str.maketrans('', '', string.punctuation))
print(f"Text without punctuations:\n{text_without_punctuations}")
```

**Plot the distribution of stopwords used in the text using barplot.**

```python
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS
from collections import Counter

# Extracting stopwords from the text
stopwords_used = [word.lower() for word in tokens if word.lower() in STOPWORDS]

# Plotting the distribution of stopwords with increased figure size
plt.figure(figsize=(15, 6))  # Adjust the size as needed
stopwords_counter = Counter(stopwords_used)
stopwords_labels, stopwords_counts = zip(*stopwords_counter.items())

plt.bar(stopwords_labels, stopwords_counts)
plt.xlabel('Stopwords')
plt.ylabel('Frequency')
plt.title('Distribution of Stopwords in the Text')
plt.show()
```

**Remove the stopwords from the text.**

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

# Download stopwords from NLTK
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Remove stopwords
stop_words = set(stopwords.words('english'))
tokens = word_tokenize(text)
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalnum()]

# Reconstruct the text without stopwords
text_without_stopwords = ' '.join(filtered_tokens)
print(f"Text without stopwords:\n{text_without_stopwords}")
```

**Plot the distribution of each POS Tag using a barplot.**

```python
import nltk
from nltk import pos_tag
from collections import Counter

# Download POS tagger from NLTK
nltk.download('averaged_perceptron_tagger')

# POS tagging
pos_tags = pos_tag(filtered_tokens)

# Plotting the distribution of POS tags
pos_tags_counter = Counter(tag for word, tag in pos_tags)
pos_tags_labels, pos_tags_counts = zip(*pos_tags_counter.items())

plt.figure(figsize=(12, 6))
plt.bar(pos_tags_labels, pos_tags_counts)
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.title('Distribution of POS Tags in the Text')
plt.show()
```

**Determine the number of unique lemma available in the text.**

```python
from nltk.stem import WordNetLemmatizer

# Download WordNetLemmatizer from NLTK
nltk.download('wordnet')

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(word) for word in filtered_tokens]

# Counting unique lemmas
num_unique_lemmas = len(set(lemmas))
print(f"Number of unique lemmas: {num_unique_lemmas}")
```

**Plot the frequency distribution of words in the text. Display only the top 10(most occurring) tokens in the chart.**

```python
from nltk import FreqDist

# Frequency distribution of words
fdist = FreqDist(filtered_tokens)

# Plotting the top 10 most occurring tokens
plt.figure(figsize=(12, 6))
fdist.plot(10, title='Top 10 Most Occurring Tokens', cumulative=False)
plt.show()
```

**Find the number of unique bigrams, trigrams and quadgrams in the corpus**

```python
from nltk import ngrams

# Function to get n-grams
def extract_ngrams(tokens, n):
    return list(ngrams(tokens, n))

# Extracting bigrams, trigrams, and quadgrams
bigrams = extract_ngrams(filtered_tokens, 2)
trigrams = extract_ngrams(filtered_tokens, 3)
quadgrams = extract_ngrams(filtered_tokens, 4)

# Counting unique bigrams, trigrams, and quadgrams
num_unique_bigrams = len(set(bigrams))
num_unique_trigrams = len(set(trigrams))
num_unique_quadgrams = len(set(quadgrams))

print(f"Number of unique bigrams: {num_unique_bigrams}")
print(f"Number of unique trigrams: {num_unique_trigrams}")
print(f"Number of unique quadgrams: {num_unique_quadgrams}")
```

**Find all dates and convert them to the DD-MM-YYYY format.**

```python
from dateutil import parser
import re
# Function to convert dates to DD-MM-YYYY format
def convert_to_dd_mm_yyyy(date_str):
    try:
        date_obj = parser.parse(date_str)
        return date_obj.strftime('%d-%m-%Y')
    except ValueError:
        return None

# Extracting and converting dates
all_dates = re.findall(r'\b\d{1,2}[-/]\d{1,2}[-/]\d{4}\b|\b\d{4}[-/]\d{1,2}[-/]\d{1,2}\b', text)
converted_dates = [convert_to_dd_mm_yyyy(date) for date in all_dates]

# Filtering out None values (invalid dates)
valid_dates = [date for date in converted_dates if date is not None]

print(f"Original dates: {all_dates}")
print(f"Converted dates: {valid_dates}")
```

**Plot a distribution of the different values of year occurring in the text.**

```python
# Extracting years using regex
year_pattern = re.compile(r'\b\d{4}\b')
years = re.findall(year_pattern, text)

# Plotting the distribution of years
year_counts = Counter(years)
year_labels, year_values = zip(*year_counts.items())

plt.figure(figsize=(12, 6))
plt.bar(year_labels, year_values)
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.title('Distribution of Years in the Text')
plt.show()
```

**Determine whether the text contains any phone numbers in it. Ensure the phone numbers are valid if any. If an invalid phone number is found, remove it from the text.**

```python
# Function to validate phone numbers
def is_valid_phone_number(phone_number):
    phone_number_pattern = re.compile(r'^\d{10}$')
    return bool(re.match(phone_number_pattern, phone_number))

# Extracting phone numbers using regex
phone_number_pattern = re.compile(r'\b\d{10}\b')
phone_numbers = re.findall(phone_number_pattern, text)

# Validating and removing invalid phone numbers
valid_phone_numbers = [number for number in phone_numbers if is_valid_phone_number(number)]
text_without_invalid_phone_numbers = re.sub(r'\b\d{11,}\b', '', text)

print(f"Original phone numbers: {phone_numbers}")
print(f"Valid phone numbers: {valid_phone_numbers}")
print(f"Text without invalid phone numbers:\n{text_without_invalid_phone_numbers}")
```
